```python
import os
from IPython.display import display, Image, Markdown

plot_dir = "/content/colab_work/results/exp10/plots"

plot_files = [
      "train_reward_guards.png",
      "train_reward_guards_smoothed.png",
      "train_loss_guards.png",
      "ppo_diagnostics_guards.png",
      "train_reward_prisoner.png",
      "train_reward_prisoner_smoothed.png",
      "train_loss_prisoner.png",
      "ppo_diagnostics_prisoner.png",
      "train_eval_link_guards.png",
      "train_eval_link_prisoner.png",
      "train_eval_link_guards.png",
      "train_eval_link_prisoner.png",
      "eval_10k_returns.png",
      "train_eval_link_prisoner.png",
      "eval_10k_returns.png",
      "eval_10k_winrate.png",
      "eval_10k_returns.png",
      "eval_10k_winrate.png",
      "eval_10k_winrate.png",
      "eval_10k_outcomes.png",
      "eval_10k_episode_length.png",
  ]


for f in plot_files:
    p = os.path.join(plot_dir, f)
    if os.path.exists(p):
        display(Markdown(f"### {f}"))
        display(Image(filename=p, width=900))
    else:
        print("Missing:", p)
```

```python
import pandas as pd

csv_path = "/content/colab_work/results/exp10/log/prisoner/rollout_returns.csv"
df = pd.read_csv(csv_path)

summary = {
      "episodes": [len(df)],
      "captured": [(df["outcome"] == "captured").sum()],
      "escaped": [(df["outcome"] == "escaped").sum()],
      "timeout": [(df["outcome"] == "timeout").sum()],
      "avg_len": [df["length"].mean()],
      "avg_return_prisoner": [df["return_prisoner"].mean()],
      "avg_return_guards_mean": [df["return_guards_mean"].mean()],
  }
display(pd.DataFrame(summary))
```
